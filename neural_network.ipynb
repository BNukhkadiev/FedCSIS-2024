# Training a Neural Network

This version above incorporates regularization, batch normalization, learning rate scheduling, and early stopping. Remember, you might need to adjust hyperparameters and experiment with different architectures to get the best performance for your specific dataset.

```python

from sklearn.metrics import confusion_matrix, make_scorer
import numpy as np


def evaluate_error(preds, gt):
    # Define the cost matrix
    cost_matrix = np.array([[0, 1, 2],
                            [1, 0, 1],
                            [2, 1, 0]])

    # Calculate the confusion matrix
    conf_matrix = confusion_matrix(gt, preds)

    # Calculate the error value
    err = np.sum(conf_matrix * cost_matrix) / len(gt)

    return err

def get_scorer():
    return make_scorer(evaluate_error,greater_is_better=False)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, LearningRateScheduler
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score
import numpy as np

# Assuming you have loaded your data into X_train, X_test, y_train, and y_test
# Preprocessing steps
X_train, X_test, y_train, y_test = train_test_split(X_train, y, test_size=0.39)
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Define early stopping callback
early_stopping = EarlyStopping(patience=10, restore_best_weights=True)

# Define a learning rate scheduler
def lr_scheduler(epoch, lr):
    if epoch < 200:
        return lr
    else:
        return lr * tf.math.exp(-0.1)

# Define learning rate scheduler callback
lr_scheduler_callback = LearningRateScheduler(lr_scheduler)

# Feature Selection using SelectKBest
selector = SelectKBest(score_func=f_classif, k=20)  # Select 20 best features
X_train_selected = selector.fit_transform(X_train_scaled, y_train)
X_test_selected = selector.transform(X_test_scaled)

# Dimensionality Reduction using PCA
pca = PCA(n_components=10)  # Select 10 principal components
X_train_pca = pca.fit_transform(X_train_selected)
X_test_pca = pca.transform(X_test_selected)

# Ensure all labels are within the valid range [0, num_classes-1]
num_classes = 3
y_train_adjusted = np.clip(y_train, 0, num_classes - 1)
y_test_adjusted = np.clip(y_test, 0, num_classes - 1)

# Build the neural network model
model = Sequential([
    Dense(512, activation='relu', input_shape=(X_train_pca.shape[1],)),
    BatchNormalization(),  # Batch normalization layer
    Dropout(0.4),  # Increased dropout
    Dense(256, activation='relu'),
    BatchNormalization(),
    Dropout(0.4),
    Dense(128, activation='relu'),
    BatchNormalization(),
    Dropout(0.4),
    Dense(64, activation='relu'),
    BatchNormalization(),
    Dropout(0.4),
    Dense(32, activation='relu'),
    BatchNormalization(),
    Dropout(0.4),
    Dense(num_classes, activation='softmax')
])

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(X_train_pca, y_train_adjusted, epochs=300, batch_size=128,
                    validation_split=0.2, callbacks=[early_stopping, lr_scheduler_callback])

# Evaluate the model on test data
test_loss, test_acc = model.evaluate(X_test_pca, y_test_adjusted)
print('Test Accuracy:', test_acc)

# Make predictions
predictions = model.predict(X_test_pca)

# Convert probabilities into class labels
predictions_classes = np.argmax(predictions, axis=1)

# Ensure that only 2000 predictions are made
if len(predictions_classes) > 2000:
    predictions_classes = predictions_classes[:2000]
elif len(predictions_classes) < 2000:
    print(f"Warning: Only {len(predictions_classes)} predictions were made. Expected 2000.")


# Continue with your existing code...

# Path to save the submission file
submission_file_path = 'submission6.txt'

# Open the file in write mode
with open(submission_file_path, 'w') as f:
    # Iterate over each prediction and write it to the file
    for prediction in predictions_classes:
        f.write(f"{prediction}\n")  # Writing each prediction on a new line

print("Submission file created:", submission_file_path)
```
